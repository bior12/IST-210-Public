*Tables:*

\
**Fish:**

Attributes: name, type, age, cost, aquarium_id

\
**Fish Food:**

Attributes: quantity, type, cost

\
**Aquariums:**

Attributes:  location, type, size

\
**Pets:**

Attributes: name, type, age, cost, enclosure_id, petfood_id

\
**Pet Food:**

Attributes:quantity, type, cost

\
**Enclosures:**

Attributes: location, type, size

\
**Animal Purchases:**

Attributes:  type(pet or fish), fish/pet_id

\
**Food Purchases:**

Attributes: type(pet or fish), food_id

\
**Customers:**

Attributes: animal purchases_id, food purchases_id, loyalty points


______________________________________________________________________________________

At the base we have the customer table. This table tracks the animal purchases, food purchases, and loyalty points. As the amount of purchases increases the loyalty increases. Each of the purchases tables stores the type of animal or food, and IDs of the animal and food. The animal purchases tables, in particular, will store whether the animal is adopted or brought. Each of the animals and fish are identified by ID. They are not identified by name or type as there can be animals and fish with the same name and type. The food is identified by ID instead of type as there can be food packages of the same type but with different amounts of food. The animals and fish tables link to the food and enclosures and aquariums to organize who eats what and who lives where. Multiple animals and fish can eat the same food and live in the same area so it is a many to one relationship between animals and fish and food and enclosures and aquariums.

By organizing the tables in this fashion, it makes it very easy to keep track of the customers. Because the customers table is at the root and can be linked to all other tables, it is very easy to keep track of which pets the customers own. This allows the shop to tailor to the customer and recommend the correct animal or fish food and perhaps recommend news pets based on previous pets. As the pet shop is a commercial business, this organization method puts the customer first.
