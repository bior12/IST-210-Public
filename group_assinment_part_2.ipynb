The airport table lists basic information about the airport. Although the airport will only store one entity—our airport— it is necessary to link the different tables together. It has foreign keys to the checkbook and the operational hours to store the high level operational information of each day or month.

\
The specific details of the day-to-day operations can be traced through the foreign keys of the customers table. The customers table will store the information about each customer such as name, class, etc, and more importantly, their flight path, plane, and airline. The flight path table stores information about the path like duration, attitude, etc, and the plane used. The plane table stores details about the planes including type, fuel capacity, etc, and the airlines it belongs to. 

\
The different employee table stores the information of the employees. Separate employees tables for each profession are used as some jobs are bound to airport and some are not. Flight attendants and pilots do not always work for the same airport since they travel in their flight, but janitors just work at the same airport everyday. The employee table stores basic information about the employees such as salary, age, etc, and where they work, linking to an airport, plane, or airline.

\
Through this type of organization, it is easy to look at the big picture details of the airport and the specific details of each customer. To look at the big picture, one only has to look at the airport table its links from foreign keys to examine how well the airport performed each day. To see the details of a specific flight or customer, one can examine the customer table or the flight path table to see its exact details.
